```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import StratifiedKFold

from sklearn.neighbors import KNeighborsClassifier
#from sklearn.neighbors import RadiusNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import BernoulliNB
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
from sklearn.naive_bayes import ComplementNB
from sklearn.svm import SVC
#from sklearn.svm import NuSVC
#from sklearn.svm import LinearSVC
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
models = []

models.append(("KNN", KNeighborsClassifier()))
#models.append(("RNN", RadiusNeighborsClassifier()))
models.append(("DT", DecisionTreeClassifier()))
models.append(("BNB", BernoulliNB()))
models.append(("GNB", GaussianNB()))
models.append(("MNB", MultinomialNB()))
models.append(("CNB", ComplementNB()))
models.append(("SVC", SVC()))
#models.append(("NSVC", NuSVC()))
#models.append(("LSVC", LinearSVC()))
models.append(("ANN", MLPClassifier()))
models.append(("RF", RandomForestClassifier()))
```

```python
import time
strat_k_fold = StratifiedKFold(n_splits=10, random_state=7, shuffle=True)

names = []
scores = []
times = []

totalStart = time.time()

for name, model in models:
    start = time.time()
    score = cross_val_score(model, X, y, cv=strat_k_fold, scoring='accuracy').mean()
    end = time.time()
    times.append(end - start)
    names.append(name)
    scores.append(score)
    
totalEnd = time.time()    
    
kf_cross_val = pd.DataFrame({'Name' : names, 'Score' : scores, 'Time' : times})
print(kf_cross_val)
print("Total Time: " + str(totalEnd - totalStart))
```
